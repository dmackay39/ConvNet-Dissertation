```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import pandas as pd
#Package imports
```

```python
training_data_generator = ImageDataGenerator(rescale = 1.0/255.)
training_data_iterator = training_data_generator.flow_from_directory('FinalProject/Training_Set_Mk8(4cl)',
                                            class_mode = "categorical", color_mode = "grayscale", target_size = (37,37), batch_size = 61, seed = 43)

val_data_generator = ImageDataGenerator(rescale = 1.0/255.)
val_data_iterator = val_data_generator.flow_from_directory('FinalProject/Validation_Set_Mk8(4cl)',
                                            class_mode = "categorical", color_mode = "grayscale", target_size = (37,37), batch_size = 109, shuffle = False)
#Importing the preprocessed image data into training and validation sets.
```

```python
model = Sequential()
model.add(tf.keras.Input(shape=(37,37,1)))
model.add(tf.keras.layers.Conv2D(30,2,strides=1,activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size = (2,2), strides = (2,2)))
model.add(tf.keras.layers.Conv2D(40,7,strides=1,activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size = (2,2), strides = (1,1)))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(32, activation = 'relu'))
model.add(tf.keras.layers.Dropout(0.46))
model.add(tf.keras.layers.Dense(4,activation = "softmax"))
#Creating convolution neural network model. Softmax output as it is predicting 4 categories. 
#This was the best version of the architecture

acc = tf.keras.metrics.CategoricalAccuracy()
auc = tf.keras.metrics.AUC()
model.compile(optimizer = 'Adam', loss = tf.keras.losses.CategoricalCrossentropy(), metrics=[acc,auc])
#The typical metrics for a classification task: loss = Categorical crossentropy. Metrics = Categorical accuracy and AUC

history = model.fit(training_data_iterator,validation_data = val_data_iterator, epochs = 500)

plt.plot(history.history['loss'], label = 'loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel("Epoch")
plt.legend()
plt.show()
plt.plot(history.history['categorical_accuracy'], label = 'accuracy')
plt.plot(history.history['val_categorical_accuracy'], label = 'val_categorical_accuracy')
plt.xlabel("Epoch")
plt.legend()
plt.show()
#Outputs a plot of the learning curves.
```

```python
losses, accuracy, aucc = model.evaluate(val_data_iterator)
#Final validation loss and accuracy at the end of training
print(losses,accuracy,aucc)
```

```python
val_data_iterator.reset()
Y_pred = model.predict(val_data_iterator)
y_pred = np.argmax(Y_pred, axis=1)
y_true = val_data_iterator.classes
cm = confusion_matrix(y_true, y_pred)
#Generation of confusion matrix.
```

```python
cm_display = ConfusionMatrixDisplay(cm).plot()
#Plot of confusion matrix. Most entries are in the diagonal indicating that the model works well. 
```

```python
print(classification_report(y_true, y_pred))
#Precision, recall and F1 score of the model. Model achieves over 0.87 in each category.
```

```python
filters, biases = model.layers[0].get_weights()

# normalize filter values to 0-1 so we can visualize them
f_min, f_max = filters.min(), filters.max()
filters = (filters - f_min) / (f_max - f_min)

n_filters, ix = 30, 1
plt.figure(figsize = (20,20), dpi = 100)
for i in range(n_filters):

    f = filters[:, :, :, i]
    # plot each channel separately
    for j in range(1):
        # specify subplot and turn of axis
        ax = plt.subplot(n_filters, 5, ix)
        ax.set_xticks([])
        ax.set_yticks([])
        # plot filter channel in grayscale
        plt.imshow(f[:, :, j], cmap='gray')
        ix += 1
# show the figure
plt.show()
plt.close()
#Plots the learned filters in the first convolutional layer.
```
